```python
df['ema_aligned_up'] =    (df['Close'] > df['EMA_7']) & (df['Close'] > df['EMA_25']) & (df['EMA_7'] > df['EMA_25'])
df['pullback'] =          (df['Close'] < df['EMA_7']) & (df['Close'] > df['EMA_25']) & (df['EMA_7'] > df['EMA_25'])
df['dip'] =               (df['Close'] < df['EMA_7']) & (df['Close'] < df['EMA_25']) & (df['EMA_7'] > df['EMA_25'])
df['ema_aligned_down'] =  (df['Close'] < df['EMA_7']) & (df['Close'] < df['EMA_25']) & (df['EMA_7'] < df['EMA_25'])
df['recover'] =           (df['Close'] > df['EMA_7']) & (df['Close'] < df['EMA_25']) & (df['EMA_7'] < df['EMA_25'])
df['rebound'] =           (df['Close'] > df['EMA_7']) & (df['Close'] < df['EMA_25']) & (df['EMA_7'] > df['EMA_25'])
df['price_ma7'] =         (df['Close'] - df['EMA_7'])
df['price_ma25'] =        (df['Close'] - df['EMA_25'])
df['ma_spread'] =         (df['EMA_7'] - df['EMA_25'])
```

```python
# prompt: Using dataframe df: create dnn 2 layers with keras, using relu for all to predict y and it is time series, add decay factor

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Assuming your dataframe is named 'df'

# Prepare the data
X = df.drop(columns=['y'])  # Features, excluding date and target variable 'y'
y = df['y'] # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False) # Split data while preserving time series order

# Scale the features (important for neural networks)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


# Build the DNN model
model = keras.Sequential([
    layers.Dense(32, activation='relu', input_shape=(X_train.shape[1],)), #Input Layer
    layers.Dense(16, activation='relu'),  # Hidden Layer
    layers.Dense(1, activation='sigmoid') # Output layer, sigmoid for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) #Using binary_crossentropy for binary classification
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
#model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
# prompt: predict y and show number of % correct

# Predict on the test set
y_pred = model.predict(X_test)

# Convert probabilities to binary predictions (0 or 1)
y_pred_binary = (y_pred > 0.5).astype(int)

# Calculate the percentage of correct predictions
correct_predictions = (y_pred_binary.flatten() == y_test.values).sum()
total_predictions = len(y_test)
percentage_correct = (correct_predictions / total_predictions) * 100

print(f"Predicted values: \n{y_pred_binary.flatten()}")
print(f"Percentage of correct predictions: {percentage_correct:.2f}%")
```

```python
X = df_sig.drop(columns=['y'])  # Features, excluding date and target variable 'y'
y = df_sig['y'] # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False) # Split data while preserving time series order

# Scale the features (important for neural networks)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


# Build the DNN model
model = keras.Sequential([
    layers.Dense(32, activation='relu', input_shape=(X_train.shape[1],)), #Input Layer
    layers.Dense(16, activation='relu'),  # Hidden Layer
    layers.Dense(8, activation='tanh'),  # Hidden Layer
    layers.Dense(1, activation='sigmoid') # Output layer, for binary classification
])

# Compile the model
adam = Adam(learning_rate=0.001)
model.compile(optimizer=adam, loss='binary_crossentropy', metrics=['accuracy']) #Using binary_crossentropy for binary classification
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
#model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
# prompt: predict y and show number of % correct

# Predict on the test set
y_pred = model.predict(X_test)

# Convert probabilities to binary ---> Check tanh or sigmoid first
y_pred_binary = (y_pred > 0).astype(int)

# Calculate the percentage of correct predictions
correct_predictions = (y_pred_binary.flatten() == y_test.values).sum()
total_predictions = len(y_test)
percentage_correct = (correct_predictions / total_predictions) * 100

print(f"Predicted values: \n{y_pred_binary.flatten()}")
print(f"Percentage of correct predictions: {percentage_correct:.2f}%")
```

```python
# prompt: plot chart green up arrow when y is True and y_pred is true, red up arrow when y is True but y_pred is false only since '2023-12-20', price in teal no any marker when signal is false

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and it has columns 'Date', 'Close', 'signal', 'y', and 'y_pred'
# and 'Date' is the index

# Filter data from '2023-12-20' onwards
df_filtered = df[df.index >= '2023-12-20']

plt.figure(figsize=(15, 8))

# Plot the price
plt.plot(df_filtered.index, df_filtered['Close'], color='teal', label='Price')

# Plot markers based on conditions
for i in range(len(df_filtered)):

    if df_filtered['y'].iloc[i] and df_filtered['y_pred'].iloc[i] and df_filtered['signal'].iloc[i]:
        plt.plot(df_filtered.index[i], df_filtered['Close'].iloc[i], 'g^', markersize=8)  # Green up arrow

    elif df_filtered['y'].iloc[i] and not df_filtered['y_pred'].iloc[i] and df_filtered['signal'].iloc[i]:
        plt.plot(df_filtered.index[i], df_filtered['Close'].iloc[i], 'r^', markersize=8)  # Red up arrow

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Price Chart with Signals')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
